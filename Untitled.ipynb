```python
from pathlib import Path
from IPython.display import display
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
import pickle


from src.config import RANDOM_STATE, TEST_SIZE, TARGET_COLUMN, MODELS_DIR, PROCESSED_DATA_DIR, FIGURES_DIR


# Cargar datos procesados
df = pd.read_csv(PROCESSED_DATA_DIR / "startup_data_processed.csv")
display(df.head())
print(f"Shape inicial: {df.shape}")
```

```python
# Eliminar columnas irrelevantes y problemáticas
prefijos_a_eliminar = [
    'closed', 'name', 'city', 'id', 'object', 'last', 'first', 'zip', 'Unnamed:', 'founded'
]
cols_problematicas = [col for col in df.columns if any(col.startswith(pref) for pref in prefijos_a_eliminar)]
df = df.drop(columns=cols_problematicas)

columnas_a_eliminar = [
    'Unnamed: 0', 'id', 'object_id', 'Unnamed: 6', 'state_code.1',
    'zip_code', 'city', 'name', 'closed_at',
    'founded_at', 'first_funding_at', 'last_funding_at', 'labels'
]
df = df.drop(columns=[col for col in columnas_a_eliminar if col in df.columns])

cols_baja_var = [col for col in df.columns if df[col].nunique() <= 1]
df = df.drop(columns=cols_baja_var)
print(f"Shape tras limpieza: {df.shape}")
```

```python
# Separar features y target
X = df.drop(TARGET_COLUMN, axis=1)
y = df[TARGET_COLUMN]
```

```python
# Dividir en train y test
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=TEST_SIZE, random_state=RANDOM_STATE
)
print(f"Tamaño del conjunto de entrenamiento: {X_train.shape}")
print(f"Tamaño del conjunto de test: {X_test.shape}")
```

```python
# ### Modelo base: Árbol de Decisión
from IPython.display import display
import pandas as pd

# Entrenamiento
dt_classifier = DecisionTreeClassifier(random_state=RANDOM_STATE)
dt_classifier.fit(X_train, y_train)

# Predicción
dt_train_pred = dt_classifier.predict(X_train)
dt_test_pred  = dt_classifier.predict(X_test)

# Métricas de accuracy
metrics = {
    "Train Accuracy": accuracy_score(y_train, dt_train_pred),
    "Test Accuracy":  accuracy_score(y_test, dt_test_pred)
}
display(pd.Series(metrics, name="Accuracy"))

# Reporte de clasificación como DataFrame
report_df = pd.DataFrame(
    classification_report(y_test, dt_test_pred, output_dict=True)
).T
display(report_df.style.format({"precision": "{:.2f}", "recall": "{:.2f}", "f1-score": "{:.2f}", "support": "{:.0f}"}))

# Matriz de confusión
fig, ax = plt.subplots(figsize=(8, 6))
sns.heatmap(
    confusion_matrix(y_test, dt_test_pred),
    annot=True, fmt="d", cmap="Blues", ax=ax
)
ax.set_title("Matriz de Confusión – Árbol de Decisión")
ax.set_xlabel("Predicción")
ax.set_ylabel("Valor real")

# Guardado
fig_dir = FIGURES_DIR / "model_performance"
fig_dir.mkdir(parents=True, exist_ok=True)
fig.savefig(fig_dir / "dt_confusion_matrix.png", bbox_inches="tight")
plt.show()
```

```python
# ### Modelo baseline: Regresión Logística
from IPython.display import display
import pandas as pd

# Entrenamiento
lr_classifier = LogisticRegression(random_state=RANDOM_STATE, max_iter=1000)
lr_classifier.fit(X_train, y_train)

# Predicción
lr_train_pred = lr_classifier.predict(X_train)
lr_test_pred  = lr_classifier.predict(X_test)

# Métricas de accuracy
metrics_lr = {
    "Train Accuracy": accuracy_score(y_train, lr_train_pred),
    "Test Accuracy":  accuracy_score(y_test, lr_test_pred)
}
display(pd.Series(metrics_lr, name="Accuracy"))

# Reporte de clasificación como DataFrame
report_lr_df = pd.DataFrame(
    classification_report(y_test, lr_test_pred, output_dict=True)
).T
display(
    report_lr_df.style.format({
        "precision": "{:.2f}",
        "recall":    "{:.2f}",
        "f1-score":  "{:.2f}",
        "support":   "{:.0f}"
    })
)

# Matriz de confusión
fig, ax = plt.subplots(figsize=(8, 6))
sns.heatmap(
    confusion_matrix(y_test, lr_test_pred),
    annot=True, fmt="d", cmap="Blues", ax=ax
)
ax.set_title("Matriz de Confusión – Regresión Logística")
ax.set_xlabel("Predicción")
ax.set_ylabel("Valor real")

# Guardado de la figura
fig_dir = FIGURES_DIR / "model_performance"
fig_dir.mkdir(parents=True, exist_ok=True)
fig.savefig(fig_dir / "lr_confusion_matrix.png", bbox_inches="tight")
plt.show()
```

```python
# — Comparativa de modelos baseline —

# 1) Definición de scores (asegúrate de tener estas variables)
train_scores = [
    accuracy_score(y_train, dt_train_pred),
    accuracy_score(y_train, lr_train_pred)
]
test_scores = [
    accuracy_score(y_test, dt_test_pred),
    accuracy_score(y_test, lr_test_pred)
]

# 2) DataFrame de comparativa
import pandas as pd
from IPython.display import display

comp_df = pd.DataFrame({
    "Modelo": ["Árbol de Decisión", "Regresión Logística"],
    "Train Accuracy": train_scores,
    "Test Accuracy": test_scores
}).set_index("Modelo")
display(comp_df.style.format("{:.4f}"))

# 3) Gráfico
import numpy as np
fig, ax = plt.subplots(figsize=(10, 6))
x = np.arange(len(comp_df))
width = 0.35

ax.bar(x - width/2, comp_df["Train Accuracy"], width, label="Train")
ax.bar(x + width/2, comp_df["Test Accuracy"], width, label="Test")
ax.set_xticks(x)
ax.set_xticklabels(comp_df.index, rotation=0)
ax.set_ylabel("Accuracy")
ax.set_title("Comparativa de Modelos Baseline")
ax.legend()
ax.set_ylim(0.5, 1.0)

# 4) Guardado
fig_dir = FIGURES_DIR / "model_performance"
fig_dir.mkdir(parents=True, exist_ok=True)
fig.savefig(fig_dir / "baseline_comparison.png", bbox_inches="tight")
plt.show()
```

```python
# ### Guardar modelos baseline
from pathlib import Path
from IPython.display import display
import pandas as pd

# 1) Crear carpeta para modelos
baseline_dir = MODELS_DIR / "baseline"
baseline_dir.mkdir(parents=True, exist_ok=True)

# 2) Serializar y guardar cada modelo
saved = {}
for name, clf in [
    ("decision_tree", dt_classifier),
    ("logistic_regression", lr_classifier)
]:
    path = baseline_dir / f"{name}.pkl"
    with open(path, "wb") as f:
        pickle.dump(clf, f)
    saved[name] = str(path)

# 3) Mostrar las rutas de los archivos guardados
display(pd.Series(saved, name="Ruta de modelo"))

# ✔️ Listo: modelos guardados en
print(f"Modelos baseline en: {baseline_dir}")
```
